```python
price = []
Bhk = []
location = []
area = []
Bathrooms = []
Furnish = []
for i in range(1,28):
    url = f"https://housing.com/rent/search-BO0P679xe73u28050522?page={i}"
    response = requests.get(url)
    soup = BeautifulSoup(response.text)
    for i in soup.find_all("div", class_="_csbfng _c8exct _g3qslr _fr1nms _1q73f6fq _4okugktf _1qkcc65e _j3gpidpf _4nrxyh40 T_a6707275"):
        if re.findall(r"\d*,*\d*,*\d*", i.text):
            price.append("".join(re.findall(r"\d*,*\d*,*\d*",i.text)))
        else:
            price.append(np.nan)

    for i in soup.find_all("div", class_="T_091c165f _sq1l2s _vv1q9c _ks15vq T_3d3547ab _7s5wglyw _5vy24jg8 _blas1v10 new-title"):
        if re.findall(r"\d*", i.text):
            Bhk.append("".join(re.findall(r"\d*",i.text)))
        else:
            Bhk.append(np.nan)

        if re.findall(r"in\s\w+\s*\w*,*\s*\w*", i.text):
            location.append("".join(re.findall(r"in\s\w+\s*\w*,*\s*\w*",i.text)))
        else:
            location.append(np.nan)

    for i in soup.find_all("section", class_="T_ea8ae345 _9j73ad _gnexct _9s1txw"):
        if re.findall(r"\d+\d", i.text):
            area.append("".join(re.findall(r"\d+\d",i.text)))
        else:
            area.append(np.nan)

        if re.findall(r"Bathrooms\d*", i.text):
            Bathrooms.append("".join(re.findall(r"Bathrooms\d*",i.text)))
        else:
            Bathrooms.append(np.nan)

        if re.findall(r"Furnishing(Semi|Un|Fully)", i.text):
            Furnish.append("".join(re.findall(r"Furnishing(Semi|Un|Fully)",i.text)))
        else:
            Furnish.append(np.nan)
```

```python
data = pd.DataFrame({"PRICE": price, "BHK": Bhk, "LOCATION": location,"AREA": area, "BATHROOMS": Bathrooms, "FURNISH": Furnish})
```

## Location Based Analysis

### Price Distribution : Compare Average rental across locations

```python
# Group by location and calculate average rental price
average_rent_by_location = data2.groupby('LOCATION')['PRICE'].mean().reset_index()

# Sort values by price for better visualization
average_rent_by_location = average_rent_by_location.sort_values(by='PRICE', ascending=False)
print(average_rent_by_location)

# Plot the comparison
plt.figure(figsize=(5,7))
plt.barh(average_rent_by_location['LOCATION'], average_rent_by_location['PRICE'], color='skyblue')
plt.xlabel('Average Rental Price')
plt.ylabel('LOCATION')
plt.title('Comparison of Average Rental Prices Across Locations')
plt.show()
```

### Avg Price per foot square

```python
data2['AREA'] = pd.to_numeric(data2['AREA'], errors='coerce')
```

```python
data2['price_per_sqft'] = data2['PRICE'] / data2['AREA']

avg_price_per_sqft_by_location = data2.groupby('LOCATION')['price_per_sqft'].mean().reset_index()

avg_price_per_sqft_by_location = avg_price_per_sqft_by_location.sort_values(by='price_per_sqft', ascending=False)
print(avg_price_per_sqft_by_location)

# Plot the comparison (Bar chart)
plt.figure(figsize=(10,10))
plt.barh(avg_price_per_sqft_by_location['LOCATION'], avg_price_per_sqft_by_location['price_per_sqft'], color='coral')
plt.xlabel('Average Price per Square Foot')
plt.ylabel('LOCATION')
plt.title('Average Price per Square Foot Across Locations')
plt.show()
```

### Price depends on locations
